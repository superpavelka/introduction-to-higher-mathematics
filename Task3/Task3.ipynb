### Задание 4

1)Решите систему уравнений:
$$\begin{equation*}
 \begin{cases}
   y = x^2 – 1, 
   \\
   e^x + x(1 – y) = 1.
 \end{cases}
\end{equation*}$$


```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-10,20,1000)
plt.plot(x,x**2 - 1);
plt.plot(x,(np.exp(x)+x-1)/x);
plt.axis('equal')
plt.axis([-2, 2, -1, 7])
plt.grid(True)
```

```python
x = np.linspace(-10,20,1000)
plt.plot(x,x**2 - 1);
plt.plot(x,1-(1-np.exp(x))/x);
plt.axis('equal')
plt.axis([-2, 2, 10, 20])
plt.grid(True)
```

```python
def equations(p):
    x,y = p
    return (y - x**2 + 1,np.exp(x) + x*(1-y) - 1)

x1,y1 = fsolve(equations, (-1,1))
x1,y1,
```

```python
x2,y2 = fsolve(equations, (2,2))
x2,y2
```

```python
x3,y3 = fsolve(equations, (4,10))
x3,y3
```
