### Задание 1

Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
Vector = namedtuple('Vector' , 'x y z')
v1 = Vector(10, 10, 10)
v2 = Vector(0, 0,-10)
v3 = Vector(v1.x + v2.x,v1.y + v2.y,v1.z + v2.z)
v3
```

```python
def vec_length(v):
    if type(v) == Vector:
        return (np.sqrt(v.x**2 + v.y**2 + v.z**2))
    else:    
        return None
```

```python
l1 = vec_length(v1)
l2 = vec_length(v2)
l1,l2
```

### Задание 3

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 0.6
plt.contour(X,Y,F,[0])
plt.gca().set_aspect('equal')
plt.show()
```
